```python
arctic_pkl = '/Users/Lisanne/Documents/AI4ER/Mres/ArcticTurbulence/data/processed_data/ml_ready/arctic2_1805.pkl'
arctic_df = pd.read_pickle(arctic_pkl)
```

```python
# Extract the features and target variable from the DataFrame
features = arctic_df[['S', 'depth', 'log_N2', 'latitude']].values
target = arctic_df['eps'].values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Define the kernel for Gaussian Process
kernel = RBF(length_scale=1.0)

# Create Gaussian Process Regressor with the defined kernel
gp = GaussianProcessRegressor(kernel=kernel)

# Train the Gaussian Process Regressor
gp.fit(X_train, y_train)

# Predict epsilon for the test set
y_pred = gp.predict(X_test)

# Calculate R2 score
r2 = r2_score(y_test, y_pred)

print("R2 score:", r2)
```
